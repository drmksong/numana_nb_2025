### Chap 23 Example 1 Velocity of Rocket (forward)

#### Problem Statement : 
>  a) Use forward difference approximation of the first derivative of ð‘£(ð‘¡) to calculate the acceleration at t=16s. Use a step size of âˆ†ð‘¡=2ð‘ .
>
>  b) Find the exact value of the acceleration of the rocket.
>
>  c) Calculate the absolute relative true error for part (b).


```python
Latex(""" \\begin{equation*} \\huge
      v(t) = 2000 \\ln[ {14 \\times 10^4  \\over 14 \\times 10^4 - 2100t}]-9.8t, \\quad 0 \\leq t \\leq 30
      \\end{equation*} 
      """)
```

```python
Latex(""" \\begin{equation*} \\huge
      a(t_i) = {v(t_{t+1}) - v(t_i)\\over \\Delta t }
      \\end{equation*} 
      """)
```

```python
def vel(t):
    if t <= 0:
        t = 0.0000001
    if t > 30:
        t = 30
    x = 14*10**4/(14*10**4 - 2100*t)
    return 2000*np.log(x)-9.8*t
```

```python
xi = np.array([i*0.1 for i in range(0, 300)])
yi = np.array([vel(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
# ax.set_ylim(0, 1000)
```

```python
v_18 = vel(18)
v_16 = vel(16)
a_16 = (v_18-v_16)/(18-16)
print(f'a({16}) = {a_16}') 
```

```python
xi = np.array([i*0.1 for i in range(0, 300)])
yi = np.array([vel(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)

xa = xi
ya = [a_16*(i-16)+v_16 for i in xi]
line2, = ax.plot(xa, ya, lw=2)
```

>  b) Find the exact value of the acceleration of the rocket.


```python
def acc(t):
    if t <= 0:
        t = 0.0000001
    if t > 30:
        t = 30
    return (-4040-29.4*t)/(-200+3*t)
```

```python
xi = np.array([i*0.1 for i in range(0, 300)])
yi = np.array([acc(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)

x0=16
y0=acc(16)
y1=a_16
ball = plt.Circle((x0, y0), 1,color='b')
ax.add_patch(ball)
ball2 = plt.Circle((x0, y1), 1,color='r')
ax.add_patch(ball2)

print(f'x0 = {x0}, y0 = {y0:0.5} m/s^2')
```

>  c) Calculate the absolute relative true error for part (b).

```python
error = abs(acc(16)-a_16)/acc(16)*100
print(f'error = {error:0.5}%')
```

### Chap 23 Example 3 Velocity of Rocket (central)

#### Problem Statement : 
>  a) Use central difference approximation of the first derivative of ð‘£(ð‘¡) to calculate the acceleration at ð‘¡=16ð‘ . Use a step size of âˆ†ð‘¡=2ð‘ .
>
>  b) Find the absolute relative true error for part (a).

```python
v_18 = vel(18)
v_14 = vel(14)

print(f'v(18) = {v_18:0.5} m/s')
print(f'v(14) = {v_14:0.5} m/s')

a_16_c = (v_18-v_14)/(18-14)
print(f'a({16}) = ({v_18:0.5}-{v_14:0.5})/{18-14} = {a_16_c:0.5} m/s^2')
```

```python
xi = np.array([i*0.1 for i in range(0, 300)])
yi = np.array([vel(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)

xa = xi
ya = [a_16_c*(i-16)+v_16 for i in xi]
line2, = ax.plot(xa, ya, lw=2)
```

>  b) Find the absolute relative true error for part (a).

```python
error = abs(acc(16)-a_16_c)/acc(16)*100
print(f'error = |{acc(16):0.5}-{a_16_c:0.5} |/{acc(16):0.5} = {error:0.5}%')
```

```python
xi = np.array([i*0.1 for i in range(0, 300)])
yi = np.array([acc(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)

x0=16
y0=acc(16)
y1=a_16_c
ball = plt.Circle((x0, y0), 1,color='b')
ax.add_patch(ball)
ball2 = plt.Circle((x0, y1), 1,color='r')
ax.add_patch(ball2)

print(f'x0 = {x0}, y0 = {y0:0.5} m/s^2')
```
