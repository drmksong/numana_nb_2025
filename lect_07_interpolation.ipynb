### Example 18.1 Linear Interpolation

#### Problem Statement : 
> - Estimate the natural logarithm of 2 using linear interpolation. 
First, perform the computation by interpolating between ln 1 5 0 and ln 6 5 1.791759. 
> - Then, repeat the procedure, but use a smaller interval from ln 1 to ln 4 (1.386294). Note 
that the true value of ln 2 is 0.6931472.

```python
Latex(f""" \\begin{{equation*}} \\huge
      f(x) = \\ln(x)
      \\end{{equation*}} 
      """)
```

```python
Latex(""" \\begin{equation*} \\huge
      f_1(x) = f(x_0) + f'(x_0)(x-x_0) = f(x_0) + {f(x_1) - f(x_0)\\over x_1 - x_0 }(x-x_0)
      \\end{equation*} 
      """)
```

```python
def func(x):
    return np.log(x)
```

```python
xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
ax.set_ylim(0, 2)
```

```python
df = func(6) - func(1)
dx = 6-1
a = df/dx
print(f'df = {df}, dx = {dx}, a = {a}')

x0,y0 = 2,func(2)
x1,y1 = 2, a*(2-1)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
line2, = ax.plot(xi, a*(xi-1), lw=2)
ball = plt.Circle((x0, y0), 0.10,color='r')
ball2 = plt.Circle((x1, y1), 0.10,color='b')
ax.add_patch(ball)
ax.add_patch(ball2)
ax.set_ylim(0, 2)

```

```python
df = func(4) - func(1)
dx = 4-1
a = df/dx
print(f'df = {df}, dx = {dx}, a = {a}')

x2,y2 = 2, a*(2-1)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
line2, = ax.plot(xi, a*(xi-1), lw=2)
ball = plt.Circle((x0, y0), 0.10, color='r')
ball2 = plt.Circle((x1, y1), 0.10, color='b')
ball3 = plt.Circle((x2, y2), 0.10, color='g')
ax.add_patch(ball)
ax.add_patch(ball2)
ax.add_patch(ball3)
ax.set_ylim(0, 2)

```

### Example 18.6 Quadratic Interpolation

#### Problem Statement : 
> - Use a Lagrange interpolating polynomial of the first and second 
order to evaluate ln 2 on the basis of the data given in Example 18.2:: 
>   - x0 = 1 f(x0) = 0
>   - x1 = 4 f(x1) = 1.386294
>   - x2 = 6 f(x2) = 1.791759


```python
Latex(""" \\begin{equation*} \\huge
         f_n(x) = \\sum_{i=0}^{n} L_i(x)f(x_i), L_i(x) = \\prod_{j=0,j\\neq i}^{n} {x-x_j \\over x_i-x_j}
        \\end{equation*} 
        """)
```

```python
def L_n(x,xi,i,n):
    assert n <= len(xi)
    L = 1
    for j in range(n):
        if j != i:
            L *= (x-xi[j])/(xi[i]-xi[j])
    return L

def f_n(x,xi,yi,n):
    
    L = np.array([L_n(x,xi,i,n+1) for i in range(n+1)])
    return np.dot(L,yi[:n+1])
```

```python
xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
```

```python
xp = np.array([1,4,6])
yp = np.array([func(x) for x in xp])
```

```python
f_1_2  = f_n(2,xp,yp,1)
f_2_2  = f_n(2,xp,yp,2)
print(f_1_2,f_2_2)

xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
plt.ylim(0,3)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
line2, = ax.plot(xi, [f_n(x,xp,yp,1) for x in xi], lw=2,color='g')
line3, = ax.plot(xi, [f_n(x,xp,yp,2) for x in xi], lw=2,color='r')

ball1 = plt.Circle((2, f_1_2), 0.10, color='g')
ball2 = plt.Circle((2, f_2_2), 0.10, color='r')

ax.add_patch(ball1)
ax.add_patch(ball2)
plt.ylim(0,3)
```

### Example 18.8 First Order Spline Interpolation

#### Problem Statement : 
> - Fit the data in following Table with first-order splines. Evaluate the
function at x = 5.

Table 18.1

| Item   | x         | f(x)       |
|--------|-----------|------------|
| 0      |      3.0  | 2.5        |
| 1      |      4.5  | 1.0        |
| 2      |      7.0  | 2.5        |
| 3      |      9.0  | 0.5        |


#### Solution :

Determine the gradient of the segment where the point of evaluation is.

```python
x = [3, 4.5, 7.0, 9.0]
fx = [2.5, 1.0, 2.5, 0.5]

for i in range(len(x)-1):
    dx = x[i+1]-x[i]
    df = fx[i+1]-fx[i]
    m = df/dx
    print(f'i = {i} df = {df}, dx = {dx}, m = {m}') 
```

```python
def find_grad(x,fx,xx):
    for i in range(len(x)-1):
        if x[i] <= xx and xx <= x[i+1]:
            dx = x[i+1]-x[i]
            df = fx[i+1]-fx[i]
            m = df/dx
            return m
    return None

def eval(x,fx,xx):
    for i in range(len(x)-1):
        if x[i] <= xx and xx <= x[i+1]:
            dx = x[i+1]-x[i]
            df = fx[i+1]-fx[i]
            m = df/dx
            return fx[i]+m*(xx-x[i])
    return None
```

```python
m = find_grad(x,fx,5.0)
ev = eval(x,fx,5.0)
print(f'm = {m}, ev = {ev}')
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x, fx, lw=2,color='b')
```

```python
ev = eval(x,fx,5.0)

fig, ax = plt.subplots()
line, = ax.plot(x, fx, lw=2,color='b')
line2, = ax.plot([5,5], [0,2.5], lw=2,color='g')

ball1 = plt.Circle((5, ev), 0.10, color='g')

ax.add_patch(ball1)
```

### Example 18.9 Qudratic Spline Interpolation

#### Problem Statement : 
> - Fit quadratic splines to the same data used in Example 18.8 (Table
18.1). Use the results to estimate the value at x = 5.


Table 18.1

| Item   | x         | f(x)       |
|--------|-----------|------------|
| 0      |      3.0  | 2.5        |
| 1      |      4.5  | 1.0        |
| 2      |      7.0  | 2.5        |
| 3      |      9.0  | 0.5        |


#### Solution :

For the present problem, we have four data points and n = 3 intervals. There- fore, 3(3) = 9 unknowns must be determined. Equations (18.29) and (18.30) yield 2(3) - 2 = 4 conditions:

```python
x = [3,4.5,7.0,9.0]
fx = [2.5,1.0,2.5,0.5]
n = len(x)-1
N3 = 3*(n)
N2 = 2*(n)

mat = np.zeros([N3,N3],dtype=float)

for i in range(n):
      for j in range(3):
            mat[2*i,3*i+j] = x[i]**(2-j)
            mat[2*i+1,3*i+j] = x[i+1]**(2-j)


# mat[0,0] = 1
# for j in range(1,N3):
#       mat[0,j] = 0
# for j in range(N3):      
#       mat[N2-1,j] = 0
# print(N2)
            
for i in range(n-1):
      for j in range(2):
            mat[N2+i,j+3*(i)] = 2**(1-j)*x[i+1]**(1-j)
            mat[N2+i,j+3*(i+1)] = -2**(1-j)*x[i+1]**(1-j)

mat[N3-1,0] = 1
print(mat)

vec_b = np.zeros(N3,dtype=float)
for i in range(n):
      vec_b[2*i] = fx[i]
      vec_b[2*i+1] = fx[i+1]

print(vec_b)
# mat = np.array([[25,5,1],[64,8,1],[144,12,1]])
# vec_b = np.array([106.8,177.2,279.2])
# print(mat)
# print(vec_b)


Latex(f""" \\begin{{equation*}}
            \\begin{{bmatrix*}} 
                  \\Huge {{{mat[0,0]}}} & \\Huge {{{mat[0,1]}}} & \\Huge {{{mat[0,2]}}}\\\\[1em]
                  \\Huge {{{mat[1,0]}}} & \\Huge {{{mat[1,1]}}} & \\Huge {{{mat[1,2]}}}\\\\[1em]
                  \\Huge {{{mat[2,0]}}} & \\Huge {{{mat[2,1]}}} & \\Huge {{{mat[2,2]}}}      
            \\end{{bmatrix*}} 
            \\begin{{bmatrix*}}
                  \\Huge a_1\\\\[1.8em]
                  \\Huge a_2\\\\[1.8em]
                  \\Huge a_3
            \\end{{bmatrix*}}=
            \\begin{{bmatrix*}}
                  \\Huge {{{vec_b[0]}}}\\\\[1em]
                  \\Huge {{{vec_b[1]}}}\\\\[1em]
                  \\Huge {{{vec_b[2]}}}
            \\end{{bmatrix*}}
      \\end{{equation*}}
""")
```

```python
import scipy.linalg

P, L, U = scipy.linalg.lu(mat)
print(P)
print(L)
print(U)

R = np.matmul(L,U)
print(np.matmul(P,R))

res = scipy.linalg.solve(mat,vec_b)
for r in res:
        print(f'{r:.4f}')
```

```python
def spunc(d,a,x):
    I = -1
    for i in range(len(d)-1):
        if d[i] <= x <= d[i+1]:
            I = i
    assert I != -1
    # print(f'I = {I}, x = {x},d[I] = {d[I]}, d[I+1] = {d[I+1]}')
    A = a[3*I]
    B = a[3*I+1]
    C = a[3*I+2]
    # print(f'A = {A}, B = {B}, C = {C}')
    n = len(a)
    res = A*x**2+B*x+C

    return (res)
```

```python
a = res
d = x
lx = []
ly = []
# print(d,a,3.5)
ev = spunc(d,a,3.5)
# print(f'f({3.5}) = {ev}')

for i in range(30,91):
    lx.append(i/10.0)
    ev = spunc(d,a,i/10.0)
    ly.append(ev)
    # print(f'f({i/10.0}) = {ev}')

fig, ax = plt.subplots()
line, = ax.plot(lx, ly, lw=2,color='b')
line2, = ax.plot([5,5], [0,2.5], lw=2,color='g')    
```
